```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/pokedex.csv', index_col=0)
df.head()
```

```python
df.columns
```

```python
cols_to_keep = ['name', 'generation',
       'status', 'species', 'type_1', 'type_2', 'height_m',
       'weight_kg', 'abilities_number', 'ability_1', 'ability_2',
       'total_points', 'hp', 'attack', 'defense',
       'sp_attack', 'sp_defense', 'speed',
       'percentage_male','against_normal',
       'against_fire', 'against_water', 'against_electric', 'against_grass',
       'against_ice', 'against_fight', 'against_poison', 'against_ground',
       'against_flying', 'against_psychic', 'against_bug', 'against_rock',
       'against_ghost', 'against_dragon', 'against_dark', 'against_steel',
       'against_fairy']
```

```python
df = df[cols_to_keep]
df.shape
```

```python
df.head()
```

```python
pd.crosstab(index=df.status, columns=df.generation)
```

```python
pd.crosstab(index=df.status, columns=df.generation, margins=True, margins_name='Total')
```

```python
6/1028
```

```python
pd.crosstab(index=df.status, columns=df.generation, normalize=True)
```

```python
3/39
```

```python
pd.crosstab(index=df.status, columns=df.generation, normalize='index')
```

```python
185/192
```

```python
pd.crosstab(index=df.status, columns=df.generation, normalize='columns')
```

```python
pd.crosstab(index=df.status, columns=df.generation, values=df.weight_kg, aggfunc='max')
```

```python
pd.crosstab(index=df.status, columns=df.generation, rownames=['which_kind?'], colnames=['gen'])
```

```python
pd.crosstab(index=df.generation, columns=df.status, values=df.percentage_male, aggfunc='mean',
            dropna=False).fillna('')
```

```python
pd.crosstab(index=[df.generation, df.status], columns=[df.species], values=df.total_points,
            aggfunc='median').fillna(0)
```
