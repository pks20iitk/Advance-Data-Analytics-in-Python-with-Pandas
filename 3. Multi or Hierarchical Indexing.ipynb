###### 8. Read in the merged_stocks dataset. 
###### You'll notice that 'Date' column is an object column right now when you call the .info() method off the dataframe.
###### Make this column a datetime object. Now, make the first three columns, namely Sector, Ticker and Date columns as index columns, thus creating a multi index

```python
merged_stocks = pd.read_csv('data/merged_stocks.csv')
merged_stocks.info() # look at the Date column, it is currently an object type
```

```python
merged_stocks['Date'] = pd.to_datetime(merged_stocks['Date'])
```

```python
merged_stocks.info() # now Date column is a datetime column!
```

```python
# We can either set index in read_csv using the argument index_col=[0,1,2] or we can set them
# afterwards using set_index() method and passing in a list of the columns we wish to see as indices.
merged_stocks.set_index(['Sector', 'Ticker', 'Date'], inplace=True)
merged_stocks
```

```python
# It is very important to check if the dataframe index is sorted.
# This is how we'll go about doing it:

merged_stocks.index.is_lexsorted()

# We see that it is true,  hence we're fine.

# Otherwise, we'd have to do merged_stocks.sort_index(inplace=True) to sort it.
```
